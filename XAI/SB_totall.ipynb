```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = open('/content/SBweights_1.npz')
```

```python
a = np.load('/content/SBweights_1.npz')
aa = a['arr_0']

b = np.load('/content/SBweights_2_1.npz')
bb = b['arr_0']

c = np.load('/content/SBweights_2_2.npz')
cc = c['arr_0']

d = np.load('/content/SBweights_3.npz')
dd = d['arr_0']
```

```python
lst = []

for i in aa:
  lst.append(i)

for i in bb:
  lst.append(i)

for i in cc:
  lst.append(i)

for i in dd:
  lst.append(i)
```

```python
l = np.array(lst)
```

```python
my_dict = {
    'Lead I':l[:,0],
    'Lead II':l[:,1],
    'Lead III':l[:,2],
    'Lead avF':l[:,3],
    'Lead avR':l[:,4],
    'Lead avL':l[:,5],
    'Lead V1':l[:,6],
    'Lead V2':l[:,7],
    'Lead V3':l[:,8],
    'Lead V4':l[:,9],
    'Lead V5':l[:,10],
    'Lead V6':l[:,11]
}
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.boxplot(my_dict.values())
plt.axline((0,0),(11,0),color='black')
ax.set_xticklabels(my_dict.keys())
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns

fig, axes = plt.subplots(figsize=(6,6))
#sns.set(style="whitegrid")
sns.violinplot(data=l, ax = axes)
plt.axline((0,0),(11,0),color='black')
plt.xticks(rotation=90)
plt.show()
```
